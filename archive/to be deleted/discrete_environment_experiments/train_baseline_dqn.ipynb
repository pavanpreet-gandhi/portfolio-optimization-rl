```python
import pandas as pd
import time
from stable_baselines3 import DQN
from environments.discrete_env_v1 import PortfolioManagementEnv
```

```python
df = pd.read_csv('data/returns_dataset.csv', index_col='Date', parse_dates=['Date'])
ASSET_COLS = ['AXP', 'AAPL', 'BA', 'GS', 'INTC', 'JNJ', 'KO', 'NKE', 'PG', 'DIS']
FEATURE_COLS = []
env = PortfolioManagementEnv(df, ASSET_COLS, FEATURE_COLS)
```

```python
directory = 'experiments/discrete_environment'

start_time_id = int(time.time()) # Seconds since January 1, 1970, 00:00:00 (UTC)
model_id = f'DQN_BASELINE_{start_time_id}'

models_dir = f'{directory}/models/{model_id}'
log_dir = f'{directory}/logs'

if not os.path.exists(models_dir):
    os.makedirs(models_dir)

if not os.path.exists(log_dir):
    os.makedirs(log_dir)
```

```python
model = DQN('MlpPolicy', env, verbose=1, tensorboard_log=log_dir)
```

```python
TIMESTEPS = 10_000 # Number of timesteps between saves
for i in range(1,100):
    model.learn(total_timesteps=TIMESTEPS, reset_num_timesteps=False, tb_log_name=model_id)
    model.save(f'{models_dir}/{TIMESTEPS*i}')
```
