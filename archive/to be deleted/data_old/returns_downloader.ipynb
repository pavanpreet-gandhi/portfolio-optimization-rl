```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import yfinance as yf
```

```python
TICKERS = ['AXP', 'AAPL', 'BA', 'GS', 'INTC', 'JNJ', 'KO', 'NKE', 'PG', 'DIS']
```

```python
stock_df = yf.download(
    TICKERS, 
    start='2012-01-01', 
    end='2022-12-31'
)['Adj Close']
```

```python
stock_df.head()
```

```python
stock_df.isna().sum()
```

```python
stock_df.plot(alpha=0.8, lw=0.8);
```

```python
returns_df = stock_df.pct_change(1).dropna()
returns_df.head()
```

```python
returns_df.plot(title='Daily Returns', alpha=0.8, lw=0.2);
```

```python
(1+returns_df).cumprod().plot(alpha=0.8, lw=0.8);
```

```python
returns_df.to_csv('data/returns_dataset.csv') # save to csv
```
