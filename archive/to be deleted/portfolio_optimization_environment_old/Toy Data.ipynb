```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
TICKERS = ['AAPL', 'GOOGL', 'MSFT']
```

```python
stock_df = yf.download(TICKERS, start='2022-01-01', end='2022-12-31')['Adj Close']
```

```python
stock_df.isna().sum()
```

```python
stock_df.plot();
```

```python
returns_df = stock_df[1:]/stock_df[:-1].to_numpy()
```

```python
returns_df.plot();
```

```python
train_data = returns_df[:'2022-08-31']
train_data.tail()
```

```python
test_data = returns_df['2022-09-01':]
test_data.head()
```

```python
train_data.to_csv('train_data.csv')
test_data.to_csv('test_data.csv')
```
